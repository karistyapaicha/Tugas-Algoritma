```python
data = [1,2,3,4,5,4,3,2,1]
terbesar = data[0]
for i in range(len(data)):
  if data[i] > terbesar:
    terbesar = data[i]
print(terbesar)
```

```python
data = [1,2,3,4,5,4,3,2,1]
terbesar = data[0]
for angka in data:
  if angka > terbesar:
    terbesar = angka
print(terbesar)
```
